**Engineer new features based on existing data (e.g., combining screen time and physical activity).**



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = 'Dots Potential Customer Survey Data.xlsx'
df = pd.read_excel(file_path)

# Engineer new features
df['Total_Screen_Time'] = (
    df['Time spent watching videos/TV'] +
    df['Total Time spent working in front of screen']
)

df['Total_Physical_Activity_Time'] = (
    df['Time spent playing indoor sports'] +
    df['Time spent playing outdoor sports']
)

df['Screen_to_Activity_Ratio'] = (
    df['Total_Screen_Time'] / (df['Total_Physical_Activity_Time'] + 1e-9)
)


```

**Visualize the distribution of features using techniques like histograms and box plots.**

```python
# Set up the plotting style
sns.set_style('whitegrid')

# Create histograms and box plots for the new features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Feature Distributions', fontsize=16)

# Histogram for Total Screen Time
sns.histplot(df['Total_Screen_Time'], bins=15, kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Distribution of Total Screen Time')
axes[0, 0].set_xlabel('Total Screen Time')

# Box plot for Total Screen Time
sns.boxplot(x=df['Total_Screen_Time'], ax=axes[0, 1], color='blue')
axes[0, 1].set_title('Box Plot of Total Screen Time')

# Histogram for Total Physical Activity
sns.histplot(df['Total_Physical_Activity_Time'], bins=15, kde=True, ax=axes[1, 0], color='green')
axes[1, 0].set_title('Distribution of Total Physical Activity')
axes[1, 0].set_xlabel('Total Physical Activity')

# Box plot for Total Physical Activity
sns.boxplot(x=df['Total_Physical_Activity_Time'], ax=axes[1, 1], color='green')
axes[1, 1].set_title('Box Plot of Total Physical Activity')


plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Create visualizations to explore relationships between features and the target variable (e.g., scatter plots).**

```python
# Scatter plots for relationships
sns.pairplot(df, vars=['Total_Screen_Time', 'Total_Physical_Activity_Time'], hue='Wear Specs', palette="Set2")
plt.suptitle("Feature Relationships with Target Variable (Wear Specs)", y=1.02)
plt.show()
```

**Perform correlation analysis to identify highly correlated features.**

```python
# Correlation matrix
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```
